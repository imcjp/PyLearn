# Jupyter Notebook 示例
這是一個增加用戶輸入的 Jupyter Notebook 範例，展示了基本的資料處理、用戶交互和視覺化過程。

```python
# 匯入所需的函式庫
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# 設置 matplotlib 支援中文
matplotlib.rcParams['font.sans-serif'] = ['Microsoft JhengHei']  # 或者 'SimHei'
matplotlib.rcParams['axes.unicode_minus'] = False  # 解決負號顯示問題

# 用戶輸入數據
print('請輸入 4 組數字，分別對應 A, B, C 三個列：')
A = []
B = []
C = []
for i in range(4):
    a = int(input(f'輸入第{i+1}個 A 值：'))
    b = float(input(f'輸入第{i+1}個 B 值：'))
    c = float(input(f'輸入第{i+1}個 C 值：'))
    A.append(a)
    B.append(b)
    C.append(c)

# 建立 DataFrame
data = {'A': A, 'B': B, 'C': C}
df = pd.DataFrame(data)
df
```

## 資料摘要
接下來，讓我們檢查 DataFrame 的摘要統計資料。

```python
# 顯示摘要統計資料
df.describe()
```

## 資料視覺化
接下來，我們將使用 matplotlib 繪製一個簡單的折線圖來視覺化資料。

```python
# 繪製折線圖
df.plot(x='A', y=['B', 'C'], kind='line')
plt.title('折線圖範例')
plt.xlabel('A')
plt.ylabel('數值')
plt.grid(True)
plt.show()
```
